```python
from pyspark.sql import SQLContext
from pyspark.sql.types import *

sqlContext = SQLContext(sc)
```

```python
sqlContext
```

```python
df = sqlContext.read.load("gs://kueppers/order_items.csv",
                         format="com.databricks.spark.csv",
                         header="true",
                         inferSchema="true")
```

```python
df.columns
```

```python
df.filter(df.quantity > 100).count()
```

```python
df.groupBy("product_id").avg("quantity").toPandas()
```

```python
df.registerTempTable( "order_items" )
```

```python
query = "Select product_id, sum(quantity), avg(quantity) from order_items group by product_id"
sqlContext.sql(query).toPandas()
```
