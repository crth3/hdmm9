```python
from kafka import KafkaConsumer
from random import randint
import datetime
import time
import sys
```

```python
KAFKA_URI = "localhost:9092"
SENSOR_TEMPERATURE_TOPIC = "sensor_temperatures"
CONSUMPTION_DELAY_IN_SECONDS = 1

consumer = KafkaConsumer(SENSOR_TEMPERATURE_TOPIC, bootstrap_servers=KAFKA_URI, 
                         auto_offset_reset='earliest',
                         #auto_offset_reset='latest',
                        )
```

```python
sensor_count = {}
try:
    for message in consumer:
        print(message.value)
        sensor = str(message.value).split(": ")[1].split(" - ")[0]
        if (sensor in sensor_count.keys()):
            sensor_count[sensor] += 1
        else:
            sensor_count[sensor] = 1
        print(sensor_count)
                
except KeyboardInterrupt:
    sys.exit()
```
